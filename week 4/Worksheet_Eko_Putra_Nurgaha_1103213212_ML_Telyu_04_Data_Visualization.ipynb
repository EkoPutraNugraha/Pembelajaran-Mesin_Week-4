Beri sample kode python untuk dapat menampilkan persentase dari data bertipe kategorikal dalam bentuk pie chart, sehingga saya bisa copy paste ke google colab.

````python
import pandas as pd
import matplotlib.pyplot as plt

# Membaca dataset
file_path = '/content/drive/MyDrive/StudentsPerformance.csv'
df = pd.read_csv(file_path)

# Membuat pie chart untuk kolom bertipe kategorikal (misalnya: 'gender')
category_column = 'gender'  # Ganti ini dengan kolom kategorikal yang ingin ditampilkan

# Menghitung persentase untuk setiap kategori
category_counts = df[category_column].value_counts()
category_labels = category_counts.index

# Membuat pie chart
plt.figure(figsize=(6, 6))
plt.pie(category_counts, labels=category_labels, autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightcoral'])
plt.title(f'Distribution of {category_column}')
plt.axis('equal')  # Agar pie chart berbentuk lingkaran
plt.show()
````

````python
import matplotlib.pyplot as plt

def plot_pie_chart(column_name):

    data = df[column_name].value_counts()
    labels = data.index
    sizes = data.values

    # Menggunakan palet warna yang aesthetic
    colors = plt.get_cmap('Set3')(range(len(labels)))

    plt.figure(figsize=(7,7))
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors)
    plt.axis('equal')  # Untuk memastikan pie chart berbentuk lingkaran
    plt.title(f'Percentage of {column_name}')
    plt.show()

plot_pie_chart('gender')
plot_pie_chart('race/ethnicity')
plot_pie_chart('lunch')
plot_pie_chart('parental level of education')
plot_pie_chart('test preparation course')
````

````python
import matplotlib.pyplot as plt

def plot_pie_chart(column_name):

    data = df[column_name].value_counts()
    labels = data.index
    sizes = data.values

    # Menggunakan palet warna hangat
    warm_colors = ['#FF7F50', '#FF6347', '#FF4500', '#FF8C00', '#FFA07A', '#FFD700', '#FFA500']

    plt.figure(figsize=(7,7))
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=warm_colors[:len(labels)])
    plt.axis('equal')  # Untuk memastikan pie chart berbentuk lingkaran
    plt.title(f'Percentage of {column_name}')
    plt.show()

plot_pie_chart('gender')
plot_pie_chart('race/ethnicity')
plot_pie_chart('lunch')
plot_pie_chart('parental level of education')
plot_pie_chart('test preparation course')
````

Prompt:

Beri kode python untuk melihat jumlah siswa yang fail dan pass untuk masing-masing pelajaran (math, reading, writing score). Dengan minimal nilai yang dianggap pass adalah 75!

````python
# Menambahkan kolom baru untuk status pass/fail berdasarkan nilai minimal 75
df['math_status'] = df['math score'].apply(lambda x: 'Pass' if x >= 75 else 'Fail')
df['reading_status'] = df['reading score'].apply(lambda x: 'Pass' if x >= 75 else 'Fail')
df['writing_status'] = df['writing score'].apply(lambda x: 'Pass' if x >= 75 else 'Fail')

# Menghitung jumlah siswa yang pass dan fail untuk masing-masing pelajaran
math_pass_fail = df['math_status'].value_counts()
reading_pass_fail = df['reading_status'].value_counts()
writing_pass_fail = df['writing_status'].value_counts()

# Menampilkan hasilnya
print("Math Pass/Fail Counts:")
print(math_pass_fail)
print("\nReading Pass/Fail Counts:")
print(reading_pass_fail)
print("\nWriting Pass/Fail Counts:")
print(writing_pass_fail)
````

Prompt:

Beri kode python yang menunjukkan visualisasi nilai math, reading, dan writing siswa, dengan melihat pengaruh dari gender, race/ethnicity, parental level of education, lunch, test preparation course-nya!

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Membaca dataset
file_path = '/content/drive/MyDrive/StudentsPerformance.csv'
df = pd.read_csv(file_path)

# Mengatur ukuran figure
plt.figure(figsize=(18, 12))

# 1. Pengaruh gender terhadap nilai math, reading, dan writing
plt.subplot(3, 3, 1)
sns.boxplot(x='gender', y='math score', data=df, hue='gender', palette='Set2', dodge=False)
plt.title('Math Score by Gender')

plt.subplot(3, 3, 2)
sns.boxplot(x='gender', y='reading score', data=df, hue='gender', palette='Set2', dodge=False)
plt.title('Reading Score by Gender')

plt.subplot(3, 3, 3)
sns.boxplot(x='gender', y='writing score', data=df, hue='gender', palette='Set2', dodge=False)
plt.title('Writing Score by Gender')

# 2. Pengaruh race/ethnicity terhadap nilai math, reading, dan writing
plt.subplot(3, 3, 4)
sns.boxplot(x='race/ethnicity', y='math score', data=df, hue='race/ethnicity', palette='Set2', dodge=False)
plt.title('Math Score by Race/Ethnicity')

plt.subplot(3, 3, 5)
sns.boxplot(x='race/ethnicity', y='reading score', data=df, hue='race/ethnicity', palette='Set2', dodge=False)
plt.title('Reading Score by Race/Ethnicity')

plt.subplot(3, 3, 6)
sns.boxplot(x='race/ethnicity', y='writing score', data=df, hue='race/ethnicity', palette='Set2', dodge=False)
plt.title('Writing Score by Race/Ethnicity')

# 3. Pengaruh parental level of education terhadap nilai math, reading, dan writing
plt.subplot(3, 3, 7)
sns.boxplot(x='parental level of education', y='math score', data=df, hue='parental level of education', palette='Set2', dodge=False)
plt.xticks(rotation=45)
plt.title('Math Score by Parental Level of Education')

plt.subplot(3, 3, 8)
sns.boxplot(x='parental level of education', y='reading score', data=df, hue='parental level of education', palette='Set2', dodge=False)
plt.xticks(rotation=45)
plt.title('Reading Score by Parental Level of Education')

plt.subplot(3, 3, 9)
sns.boxplot(x='parental level of education', y='writing score', data=df, hue='parental level of education', palette='Set2', dodge=False)
plt.xticks(rotation=45)
plt.title('Writing Score by Parental Level of Education')

plt.tight_layout()
plt.show()
````

Prompt: Beri kode untuk melihat nilai rata-rata seluruh mata pelajaran (math, reading writing) dari setiap siswa dan lihat korelasinya dengan masing-masing kategori

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Membaca dataset
file_path = '/content/drive/MyDrive/StudentsPerformance.csv'
df = pd.read_csv(file_path)

# Menambahkan kolom baru untuk nilai rata-rata
df['average score'] = df[['math score', 'reading score', 'writing score']].mean(axis=1)

# Visualisasi rata-rata nilai berdasarkan kategori
plt.figure(figsize=(18, 12))

# 1. Rata-rata nilai berdasarkan gender
plt.subplot(3, 2, 1)
sns.boxplot(x='gender', y='average score', data=df, hue='gender')
plt.title('Average Score by Gender')

# 2. Rata-rata nilai berdasarkan race/ethnicity
plt.subplot(3, 2, 2)
sns.boxplot(x='race/ethnicity', y='average score', data=df, hue='race/ethnicity')
plt.title('Average Score by Race/Ethnicity')

# 3. Rata-rata nilai berdasarkan parental level of education
plt.subplot(3, 2, 3)
sns.boxplot(x='parental level of education', y='average score', data=df, hue='parental level of education')
plt.xticks(rotation=45)
plt.title('Average Score by Parental Level of Education')

# 4. Rata-rata nilai berdasarkan lunch
plt.subplot(3, 2, 4)
sns.boxplot(x='lunch', y='average score', data=df, hue='lunch')
plt.title('Average Score by Lunch')

# 5. Rata-rata nilai berdasarkan test preparation course
plt.subplot(3, 2, 5)
sns.boxplot(x='test preparation course', y='average score', data=df, hue='test preparation course')
plt.title('Average Score by Test Preparation Course')

plt.tight_layout()
plt.show()

# Korelasi antar variabel kategorikal dan average score
correlation_data = df[['average score']]
print(correlation_data.corr())
````
